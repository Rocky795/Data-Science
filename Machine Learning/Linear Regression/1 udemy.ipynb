```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./Advertising.csv')
```

```python
df.head()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
x=df.drop('sales',axis=1)
```

```python
x
```

```python
y=df['sales']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.3,random_state=101
)
```

```python
len(df)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

# Regression Performance Evaluation

```python
x_test.head()
```

```python
y_test.head()
```

```python
test_prediction=model.predict(x_test)
```

```python
test_prediction
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['sales'].mean()
```

```python
sns.histplot(data=df,x='sales',bins=20)
```

```python
mean_absolute_error(y_test,test_prediction)
```

```python
mean_squared_error(y_test,test_prediction)
```

```python
np.sqrt(mean_squared_error(y_test,test_prediction))
```

```python
test_reseduals=y_test- test_prediction
```

```python
test_reseduals
```

```python
sns.scatterplot(x=y_test,y=test_reseduals)
plt.axhline(y=0,color='r',ls='--')
# since graph look random then there is no auto corelation of errors
```

```python
sns.displot(data=test_reseduals,bins= 25,kde=True)
# Homoscedesticity
```

# Residual Error

```python
RSS=np.mean(test_reseduals**2)
RSS
```

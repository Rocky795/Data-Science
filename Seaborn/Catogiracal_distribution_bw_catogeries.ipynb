```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.boxplot(data=df,y='math score')
```

```python
sns.boxplot(data=df,y='math score',x='test preparation course')
```

```python
sns.boxplot(data=df,y='reading score',x='test preparation course')
```

```python
plt.figure(figsize=(12,5),dpi=150)
sns.boxplot(data=df,y='reading score',x='parental level of education')
```

```python
plt.figure(figsize=(12,5),dpi=150)
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='test preparation course')
```

```python
plt.figure(figsize=(12,5),dpi=150)
sns.violinplot(data=df,y='reading score',x='parental level of education',hue='test preparation course',
inner='quartile'
)
```

```python
plt.figure(figsize=(12,5),dpi=150)
sns.violinplot(data=df,y='reading score',x='parental level of education',hue='test preparation course',split=True,
inner='quartile'
)
```

```python
sns.swarmplot(data=df,x='math score',y='gender',s=3)
```

```python
sns.boxenplot(x='math score',y='test preparation course',data=df)
```

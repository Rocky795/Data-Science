```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('./dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue="level of education",palette='Dark2',s=100,alpha=0.5)
```

```python
plt.figure(figsize=(12,9),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue="level of education",palette='Dark2',size='work experience',style='level of education'
)
plt.savefig('Sales-Salary.jpg')
```

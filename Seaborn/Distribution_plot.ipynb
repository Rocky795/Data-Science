```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,5),dpi=150)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
# Do not use Distplot it will removed in future
sns.displot(data=df,x='salary',bins=20,kde=True,rug=True)
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='blue',linewidth=4,linestyle='--')
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.kdeplot(data=df,x='salary')
```

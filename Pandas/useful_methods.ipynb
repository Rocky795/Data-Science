```python
import pandas as pd
```

```python
df=pd.read_csv('./tips.csv')
```

```python
df.head()
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(1234567)
```

```python
df['CC Number'].apply(last_four)
```

```python
df['last_four']=df['CC Number'].apply(last_four)
df
```

```python
def yelp(num):
    if num<10:
        return '$'
    elif num<30:
        return "$$"
    else:
        return "$$$"
```

```python
df['yelp']=df['total_bill'].apply(yelp)
df
```
